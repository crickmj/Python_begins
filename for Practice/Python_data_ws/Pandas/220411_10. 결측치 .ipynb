## 데이터 제외하기 dropna

```python
import pandas as pd

df = pd.read_csv("score.csv", index_col='지원번호')
df
```

```python
df.dropna(inplace=True) # 전체 데이터 중에서 NaN은 무조건 삭제 --> 열이 날라가네.. R와 같은듯 
```

```python
df
```

```python
df = pd.read_csv("score.csv", index_col='지원번호')
df
```

- axis : index 또는 columns  : row를 지울 지, col을 지울 지 
- how : any 또는 all : 하나라도 NaN이 있으면 삭제하거나, 아니면 전체가 NaN이어야 삭제 

```python
df.dropna(axis='index', how='any') # NaN이 하나라도 있는 row 삭제 
```

```python
df.dropna(axis='columns', how='any') # how는 any가 디폴트
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # column 데이터 전체가 NaN인 경우에만 삭제
```
