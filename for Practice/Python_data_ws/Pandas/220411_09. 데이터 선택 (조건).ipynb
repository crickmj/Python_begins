# 09. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd

df = pd.read_csv("score.csv", index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인 지 여부를 True / False
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # filt 아닌 것 적용 
```

```python
df[df['키'] >= 185] # filt = (df['키'] >= 185), df[filt] 와 동일 
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 점수 df.loc[row_selection , col_selection]
```

```python
df.loc[df['키'] >= 185, ['이름','수학','과학']] # 키 185 이상 중, 이름/수학/과학 가져옴 
```

## 다양한 조건 

& 조건 (그리고)

```python
df.loc[(df['키']>=185) & (df['학교'] == '북산고')] # 키가 185 이상인 북산고 학생들 데이터
```

| 또는 조건 

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키 170 미만 또는 200 초과인 학생들
```

str 함수

```python
filt = df['이름'].str.startswith('송') # 송으로 시작하는 것 
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 태가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름이 '태' 들어가는 사람 제외 
```

```python
langs = ['Python','Java'] 
filt = df['SW특기'].isin(langs) # software 특기가 파이썬/자바인 사람 
df[filt]
```

```python
langs = ['python','java'] 
filt = df['SW특기'].str.lower().isin(langs) # 다 포함! 
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt] # NaN 데이터 안됨 
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터 전체 true
```

```python
filt = df['SW특기'].str.contains('Java', na=False) # False 처리해야함
df[filt]
```
