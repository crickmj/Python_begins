# 12. 데이터 수정 

```python
import pandas as pd

df = pd.read_csv("score.csv", index_col='지원번호')
df
```

## column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고' : '무슨고'}) # dictionary  형태로! 
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True) # dictionary  형태로! 
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower() 
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + '등학교' 추가됨 
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' #  결과 column을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 결과에 대해 총합이 400보다 큰 애들이 Pass
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 컬럼 삭제
```

```python
df.drop(columns=['국어','영어','수학'])
```

## row 삭제

```python
df.drop(index="4번") # 4번 row 삭제
```

```python
filt = df['수학'] < 80 # 80점 미만 수학..
df[filt]
```

```python
df.drop(index=df[filt].index)
```

## row 추가

```python
df.loc['9번'] = ['이정환','해남고등학교',184,90,90,90,90,90,'Kotlin',450,'Pass'] # 새로운 row 추가
df
```

## cell 수정 (특정)

```python
df.loc['4번', 'SW특기'] = 'Python' #4번 학생 software 특기 변경
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 맨 뒤 결과 컬럼 (-1) 을 제일 앞, 나머지는 순서대로 !
df
```
