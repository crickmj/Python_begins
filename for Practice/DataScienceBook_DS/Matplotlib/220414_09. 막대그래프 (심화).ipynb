# 09. 막대그래프 (심화)

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 

plt.barh(labels, values) # bar + h  수평으로 변경 
```

```python
plt.barh(labels, values)
plt.xlim(175, 195) # 수평으로 바꼈으니까 xlim 
```

```python
bar = plt.bar(labels, values)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('.') # ///// 
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color = 'steelblue')
```
