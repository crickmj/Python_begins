# 03. 범례 (legend)

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 

x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, label="무슨데이터") # label = "레전드 명"
plt.legend() 
```

```python
plt.plot(x,y, label="무슨데이터")
plt.legend(loc='upper right') 
```

```python
plt.plot(x,y, label="무슨데이터")
plt.legend(loc='lower right') # loc 속성이 여러개 있음 matplotlib.org 에서 확인 가능. best가 default이고, 데이터 최적으로 된 위치에 기재 
```

```python
plt.plot(x,y, label="범례")
plt.legend(loc=(0.5, 0.5)) # x축, y축 기준으로 [0 - 1] 사이, tuple로 넣어주면 됨
```
