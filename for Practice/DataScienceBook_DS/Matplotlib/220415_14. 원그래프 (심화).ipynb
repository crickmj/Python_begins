# 14. 원 그래프 (심화)

```python
import pandas as pd

import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'etc']
colors = ['#ffadad','#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05] * 6

plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors,
        explode = explode)
plt.show()
```

```python
wedgeprops = {'width' : 0.6} # 도넛 형태로 함!! 
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors,
        wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops = {'width' : 0.6, 'edgecolor': 'w', 'linewidth': 2} # 도넛 형태로 함!! 
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors,
        wedgeprops=wedgeprops)
plt.show()
```

```python
# 10% 미만인 값은 보여주지 않도록 함수 설정

def custom_autopct(pct):
    # return ('%.1f%%' % pct) if pct > 10 else ''
    # return '{:.1f}%'.format(pct) if pct >= 10 else '' 
    return '{:.0f}%'.format(pct) if pct >= 10 else '' # 정수 보여주도록 0f 

plt.pie(values, labels = labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors,
        wedgeprops=wedgeprops, pctdistance= 0.7) # pctdistance = 클 수록 가운데로부터 멀어짐 
plt.show()
```

## Dataframe 활용

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv')
df
```

```python
# 학교별로 학생이 몇 명 인지!

grp = df.groupby('학교')
grp.size() # or grp.size()['능남고']
```

```python
values = [grp.size()['북산고'], grp.size()['능남고']]
labels = ['북산고','능남고']
clrs = ['r','steelblue']
wedgeprops = {'width' : 0.5, 'edgecolor': 'w', 'linewidth' : 1}
explode = [0.1, 0.1]
 
plt.pie(values, labels = labels, autopct='%.1f%%', colors = clrs, wedgeprops = wedgeprops, explode = explode,
        pctdistance = 0.7)
plt.title('소속 학교')
plt.legend(loc=(0.5, 0), ncol=2)
plt.show()
```
