# 16. 여러 그래프

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv')
df
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,10)) # 하위 플랏을 몇 개 그릴건지? 2x2로 4개! 생성
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0,0].bar(df['이름'], df['국어'], label = '국어점수')
axs[0,0].set_title('첫 번째 그래프')
axs[0,0].set(xlabel='이름', ylabel='점수') # x,y 축 label 설정
axs[0,0].legend()
axs[0,0].set_facecolor('lightyellow') #전경 색 
axs[0,0].grid(linestyle=':', linewidth=0.5)

# 두 번째 그래프

axs[0,1].plot(df['이름'], df['수학'])
axs[0,1].plot(df['이름'], df['영어'])
axs[0,1].set_title('두 번째 그래프')
axs[0,1].set(xlabel='이름', ylabel='수학,영어')
axs[0,1].legend()

# 세 번째 그래프

axs[1,0].barh(df['이름'], df['키'])
axs[1,0].legend()

# 네 번째 그래프
 
axs[1,1].plot(df['이름'], df['사회'], color='green', alpha=0.5)
axs[1,1].legend()
```
