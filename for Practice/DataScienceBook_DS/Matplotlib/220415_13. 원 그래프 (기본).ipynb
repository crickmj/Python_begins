# 13. 원 그래프 (기본)

```python
import pandas as pd

import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'etc']

plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False) 
# pie chart는 label = xx 아니고 labels! , '%.1f' %로 소수점 하나. 뒤에 붙히면 % 단위, counterclock으로 어느 방향으로!? 
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'etc']
# explode = [0.2, 0.1, 0, 0, 0, 0] # 안띄우려면 0 으로 써야함 ! 
explode = [0.1] * 6 # 0.5 를 6개 씩 
plt.pie(values, labels = labels, explode=explode, autopct='%.1f')
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode, autopct = '%.1f%%', labeldistance=1)
plt.legend(loc=(1.2, 0.3)) # loc 튜플로 
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode, autopct = '%.1f%%', labeldistance=1)
plt.title('언어 별 선호도')
plt.legend(loc=(1.2, 0.3), title='언어 별 선호도') # legend에 title 가능
plt.show()
```
