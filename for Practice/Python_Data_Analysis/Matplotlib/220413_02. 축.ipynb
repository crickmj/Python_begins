# 02. 축 

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title("꺾은선 그래프", fontdict={'family' : 'AppleGothic', 'size' : 15})
```

```python
plt.plot(x,y)
plt.xlabel("x축", color="red")
plt.ylabel("y축", color="#00aa00")  # rgb 사용 
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("x축", color="red", loc = 'right') # right, left, center
plt.ylabel("y축", color="#00aa00", loc = 'top') # top, bottom, center 
plt.show()
```

```python
plt.plot(x,y)
plt.xticks([1,2,3,4])
plt.yticks([3,6,9,12])
plt.xlabel("x축", color="red", loc = 'right') # right, left, center
plt.ylabel("y축", color="#00aa00", loc = 'top') # top, bottom, center 
plt.show()
```
