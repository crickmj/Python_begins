# 08. 막대그래프 (기본)

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 

plt.bar(labels, values) 
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 
colors = ['r','g','b']

plt.bar(labels, values, color=colors, alpha = 0.5) 
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 

plt.bar(labels, values)
plt.ylim(175, 195) # plt.ylim (x,y) x - y까지
```

```python
plt.bar(labels, values, width=0.5) # 두께 변경
```

```python
plt.bar(labels, values, width=0.4)
plt.xticks(rotation = 45) # x 축 이름 값 각도 45도로 설정
plt.yticks(rotation = 45)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184] 
ticks = ['1번 학생','2번 학생','3번 학생']

plt.bar(labels, values)
plt.ylim(175, 195)
plt.xticks(labels,ticks, rotation=45) # x tick을 labels에서 틱으로! 
```
