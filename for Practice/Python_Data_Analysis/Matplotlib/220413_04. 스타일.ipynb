# 4. 스타일

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기 
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용할 때 - 가 깨지는 현상 해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, label="practice")
plt.legend()
```

```python
plt.plot(x,y, linewidth=5) # line 두께 
```

## 마커

```python
plt.plot(x,y, marker='o') # point 찍어줌
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='o', markersize=10, linestyle='None')
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize = 10, markerfacecolor='black', markeredgecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

## 색 정의

```python
plt.plot(x,y, linestyle = '--', color='black', linewidth=1, marker = 'o', markeredgecolor = 'black',
         markerfacecolor = 'yellow', markersize = 9) # 심심해서 다 넣어봤다.. 
```

# 포맷

```python
plt.plot(x,y, 'ro--') #'색/마커스타일/라인스타일' 'color , Marker, linestyle' 
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y, 'go') # line style 없으면 선 안나타남
```

# 축약어
markeredgecolor = mec, markerfacecolor = mfc, linestyle = ls, 등 matplotlib.org에서 확인 가능

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', ls=':') # 축약어
```

# 투명도

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, alpha = 0.3) # alpha : 투명도 (0-1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5)) # width = 10, heigh = 5
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10)) 
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10), dpi=200) # inch 당 dot 수 = 해상도 
plt.plot(x,y, 'bv:', label = 'practice')
plt.legend()
```

## background color

```python
plt.figure(facecolor='#a1c2fd')

plt.plot(x,y, 'bo:', label='practice')
plt.title("Practice complex")

plt.legend()
plt.show()
```
