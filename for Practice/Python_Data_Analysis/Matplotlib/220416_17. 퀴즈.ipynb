# Matplotlib 퀴즈

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    '영화': ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도': [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수': [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], #단위 만명
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10] 
}

df = pd.DataFrame(data)
df
```

## 1번 문제. 
영화 데이터를 활용하여, x 축은 영화, y축은 평점인 막대 그래프를 만드시오 

```python
plt.bar(df['영화'], df['평점'])
plt.show()
```

## 2번 문제
앞에서 만든 막대 그래프에서 아래 제시된 사항을 추가하시오.
 1) 제목: '국내 Top 8 영화 평점 정보'
 2) x축 레이블: 영화 (90도 회전)
 3) y축 레이블: 평점

```python
plt.bar(df['영화'], df['평점'])
plt.title('국내 Top 8 영화 평점 정보')
plt.xlabel('영화')
plt.xticks(rotation=90)
plt.ylabel('평점')
plt.show()
```

## 3번 문제 
개봉 연도 별 평점 변화 추이를 꺾은선 그래프로 나타내시오.

```python
df_group = df.groupby('개봉 연도').mean()
df_group

plt.plot(df_group.index, df_group['평점'])
plt.title('개봉 연도 별 평점')
plt.show()
```

## 4번 문제
3번 문제의 그래프에서 제시된 세부 사항을 적용하시오.
 1) marker = 'o'
 2) x축 눈금 = 5년 단위 (2005, 2010, 2015, 2020)
 3) y축 눈금 = 최소 7, 최대 10

```python
df_group = df.groupby('개봉 연도').mean()
df_group

plt.plot(df_group.index, df_group['평점'], marker='o', ls='--', lw=0.5, color='black')
plt.title('개봉 연도 별 평점')
plt.xticks([2005, 2010, 2015, 2020], fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(7, 10)
plt.show()
```

## 5번 문제
평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 제시된 세부 사항을 적용하여 그리시오
1) label: 9점 이상 / 9점 미만
2) 퍼센트: 소수점 첫째자리 까지 표시
3) 범례: 그래프 우측에 표시

```python
filt = df['평점'] >= 9.0
df[filt] 
values = [len(df[filt]), len(df[~filt])] # 3과 5 개 나뉘어짐
labels = ['9점 이상', '9점 이하']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1,0.3))
plt.show()
```
