## Matplotlib 253 페이지

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

## 인터페이스 2개

```python
plt.subplot(2,1,1) # rows, cols, pannel number
plt.plot(x, np.sin(x))

plt.subplot(2,1,2) # rows, cols, pannel number 2줄 중에 2번째 란 뜻
plt.plot(x, np.cos(x))

# 근데 이렇게 하면.. 첫 번째 그림 수정은 ?
```

```python
# 객체지향 인터페이스

# 플랏 그리드 생성
# ax는 두 개의 축 개열 배열이 됨

fig, ax = plt.subplots(2) 

ax[0].plot(x, np.sin(x)) 
ax[1].plot(x, np.tan(x)) 
```
