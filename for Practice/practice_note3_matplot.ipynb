## 밀도 플롯과 등고선 플롯

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z, colors='black') # 양수 실선, 음수 점선
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0,5,0,5], origin = 'lower',
           cmap = 'RdGy', alpha = 0.5)
plt.colorbar()
```

```python
## 커널 밀도 측정

from scipy.stats import gaussian_kde
# 배열 크기 맞추기
data = np.vstack([x,y])
kde = gaussian_kde(data)
```

```python
# 정규 그리드 평가

xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
```

```python
## 이미지

plt.imshow(Z.reshape(Xgrid.shape), 
           origin = 'lower', aspect = 'auto',
           extent = [-3.5, 3.5, -6, 6], cmap = 'Blues')
cb = plt.colorbar()
cb.set_label('density')
```

```python
## 정규분포 데이터 만들기

mean = [0, 0]
cov = [[1,1], [1,2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# gridspec 으로 축 설정

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4, hspace = 0.2, wspace = 0.2)
main_ax = fig.add_subplot(grid[:-1, 1])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels = [], sharey = main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels = [], sharex = main_ax)

# 메인 축에 점 산포
main_ax.plot(x, y, 'ok', markersize = 3, alpha = 0.2)

# 보조축 히스토그램

x_hist.hist(x, 40, histtype = 'stepfilled',
            orientation = 'vertical', color = 'gray')
x_hist.invet_yaxis()

y_hist.hist(y, 40, histtype = 'stepfilled',
            orientation = 'giruzibtal', color = 'gray')
y_hist.invet_xaxis()
```

## 직접 플롯 변경하기

```python
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
ax = plt.axes(facecolor = '#E6E6E6') # bg color 회색
ax.set_axisbelow(True)

plt.grid(color="w", linestyle = "solid") # 그리드 하얀색으로 

for spine in ax.spines.values():
    spine.set_visible(False) # 축 선 (spine) 숨김
    
ax.xaxis.tick_bottom()
ax.yaxis.tick_left() # 왼쪽/오른쪽 눈금 숨김

ax.tick_params(colors = 'gray', direction = 'out')

for tick in ax.get_xticklabels():
    tick.set_color('gray')

for tick in ax.get_yticklabels():
    tick.set_color('gray') # 눈금과 레이블 색 밝게 변경
    

ax.hist(x, edgecolor = '#E6E6E6', color = '#EE6666')   
```

```python
def hist_and_lines():
    np.random.seed(0)
    fix, ax = plt.subplots(1, 2, figsize = (11,4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b','c'], loc = 'lower left')
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

```python
with plt.style.context('ggplot'):
    hist_and_lines()
```
